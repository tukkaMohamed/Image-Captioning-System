```python
import pandas as pd

df = pd.read_csv(
    f"{BASE}/captions.txt"
)

df.head()
```

```python
print(df.head())

print(df.shape)

print(df.isnull().sum())

print(df.describe())
```

```python
print("Before:", len(df))

df = df.drop_duplicates()

print("After:", len(df))
```

```python
import re

def clean_caption(text):
    text = text.lower()

    text = re.sub(r"[^a-zA-Z ]", "", text)

    text = re.sub(r"\s+", " ", text)

    return text.strip()

df["caption"] = df["caption"].apply(clean_caption)
```

```python
df.head()
```

```python
df["caption"] = df["caption"].apply(
    lambda x: "<start> " + x + " <end>"
)
```

```python
from sklearn.model_selection import train_test_split

images = df["image"].unique()

train_imgs, test_imgs = train_test_split(
    images,
    test_size=0.2,
    random_state=42
)

train_imgs, val_imgs = train_test_split(
    train_imgs,
    test_size=0.1,
    random_state=42
)
```

```python
train_df = df[df.image.isin(train_imgs)]

val_df = df[df.image.isin(val_imgs)]

test_df = df[df.image.isin(test_imgs)]
```

```python
print(train_df.shape)
print(val_df.shape)
print(test_df.shape)
```

```python
print("Train Images:", train_df["image"].nunique())
print("Validation Images:", val_df["image"].nunique())
print("Test Images:", test_df["image"].nunique())
```

```python
print(set(train_df["image"]).intersection(set(val_df["image"])))
print(set(train_df["image"]).intersection(set(test_df["image"])))
print(set(val_df["image"]).intersection(set(test_df["image"])))
```

```python
train_df.to_csv(f"{BASE}/dataset_caption_train.csv", index=False)
val_df.to_csv(f"{BASE}/dataset_caption_val.csv", index=False)
test_df.to_csv(f"{BASE}/dataset_caption_test.csv", index=False)
```
